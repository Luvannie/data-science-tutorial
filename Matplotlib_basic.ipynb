# Pyplot API vs Object-oriented API

### Pyplot API

```python
x=np.linspace(0,100,100)
plt.plot(x,np.sin(x),color="purple",linestyle="dashed",label="sin(x)")
plt.xlabel(" x")

plt.plot(x,np.cos(x),color="green",label="cos(x)")
plt.xlabel("bin x")
plt.ylabel("F(x)")

plt.legend() # show ra nhng nhn  dn
plt.title("2 hm lng gic sin cos")
plt.axis("tight")
plt.axis([0,100,-2,2])
```

### OO-API

```python
# prepare data
x=[1,3,6,10]
y=[1,27,216,1000]

# setup plot
fig,ax =plt.subplots(figsize=(5,5))

#plot the data
ax.plot(x,y)
ax.set(title="A simple subplot",xlabel="X",ylabel="x^3")
```

# 2. Most common type of Matplotlib plots

### Line

```python
x=np.arange(0,100,1)
# OO api
fig,ax=plt.subplots()
ax.plot(x,x**3)
```

### Scatter

```python
#pyplot api
y=[np.math.factorial(i) for i in x]
plt.scatter(x,y)
print(y)
```

```python
#oo api
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
sizes=1000*rng.rand(100)
color=1000*rng.rand(100)
fig,ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=color,cmap="flag",alpha=0.8)
fig.colorbar(img1)
```

### 3. Bar

# Data Visualization- California cities

```python
df=pd.read_csv("california_cities.csv")
```

```python
df.describe()
```

```python
df.head()
```

```python
# extract latd( v ) and longtd(kinh )
latd= df["latd"]
longd=df["longd"]

#extract area
area_total,area_land,area_water=df["area_total_km2"],df["area_land_km2"],df["area_water_km2"]
pop=df["population_total"]
# plot using pyplot
plt.figure(figsize=(15,15)) # ci ny phi vit trc th mi chnh c size
plt.scatter(longd,latd,c=np.log(pop),cmap="viridis",s=area_total,alpha=0.6)
plt.style.use("seaborn")
plt.axis("equal")
plt.xlabel("Kinh ")
plt.ylabel("V ")
plt.title("California")
plt.colorbar(label="log(population)")
area_range=[50,100,300,500,1000,1500]
for area in area_range:
    plt.scatter([],[],s=area, label=str(area)+"km$^2$")
plt.legend(scatterpoints=1,labelspacing=1,title="city area")
```
